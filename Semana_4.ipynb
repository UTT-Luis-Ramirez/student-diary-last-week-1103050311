### 2. Mediana.

Sinónimo de “Medio y  la mediana muestral ”, es en realidad el valor medio una vez estén ordenados los elementos de la población de menor a mayor.

Si la  población (*n*) es  **impar**, la mediana se calcula:

$$
x = x_{{n+1}/2}
$$

Si es par:

$$
x = x_{n/2} + x_{n/2 +1}
$$

### 3. Media Recortada.

La **media recortada** es una medida de tendencia central que se calcula eliminando un porcentaje de los valores extremos (los más bajos y los más altos) antes de calcular la media aritmética.

- Reduce el efecto de valores atípicos .
- Da una representación más estable del promedio cuando hay datos extremos.
- Se usa en estadísticas deportivas, finanzas y estudios científicos.

Se calcula:

1. Ordenando los datos de menor a mayor.
2. **Eliminando un porcentaje** de los datos **en ambos extremos** (por ejemplo, un 10% o 20%).
3. Sacando la **media aritmética** con los valores restantes.

### 4. Cuartiles.

Los cuartiles dividen un conjunto de datos en **4 partes iguales**, cada una con el **25%** de los valores.

**Q1 (Primer cuartil - 25%)**: Es el valor que deja el 25% de los datos por debajo.
**Q2 (Segundo cuartil - 50%)**: Es la **mediana** del conjunto, dejando el 50% de los datos por debajo.
**Q3 (Tercer cuartil - 75%)**: Es el valor que deja el 75% de los datos por debajo.

### 5. Percentiles.

Los percentiles dividen un conjunto de datos en **100 partes iguales**, por lo que cada percentil representa el **1%** de los datos.

 **Ejemplos de percentiles :**

 **P25  =** (Percentil 25%)

 **P40  =** (Percentil 40%)

 **P70  =** (Percentil 70%)

 Para calcular los perentiles primero se organizan los datos de menor a mayor y luego se calcula:

 $$
Posicion  P_k = \frac{(n+1)k}{{100}}
$$

Donde *k*  pertenece a {1,2,...,99}.

El **rango** mide la diferencia entre el valor **máximo** y el **mínimo** en un conjunto de datos.

$$
